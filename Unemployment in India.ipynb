# Data Preparation

```python
X = employ_df.iloc[:,3:6]
Y = employ_df.iloc[:,3]
```

```python
print(X)
```

```python
print(Y)
```

```python
sns.histplot(X, bins = 10)
plt.title("Unemployment in India")
plt.show()
```

# Train_test_split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 20)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

# Fitting the Training models

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, Y_train)
```

# Prediction

```python
y_test_predicted = model.predict(X_test)
print(y_test_predicted)
print(type(y_test_predicted))
print(Y_test)
print(type(Y_test))
```

```python
plt.hist(Y_test, bins = 20)
plt.title("Y_test")
plt.show()
```

```python
plt.hist(y_test_predicted, bins = 20)
plt.title("y_test_predicted")
plt.show()
```

The above two plottings for Y_test an y_test_predicted are exactly equal

# r2_score

```python
from sklearn.metrics import r2_score
print("r2_score is :", r2_score(Y_test, y_test_predicted)*100)
```
